# 21) Crea un DataFrame pandas da un dizionario.

```python
import pandas as pd

# Definisci un dizionario con dati
dati = {
    'Nome': ['Daniele', 'Marco', 'Chiara', 'Francesco', 'Maria'],
    'Età': [25, 27, 32, 23, 45],
    'Città': ['Roma', 'Monza', 'Bergamo', 'Firenze', 'Bari']
}

# Crea un DataFrame pandas dal dizionario
df = pd.DataFrame(dati)

# Stampa il DataFrame
print(df)
```

# 24) Filtra un DataFrame pandas in base a una condizione.

```python
import pandas as pd
import sys

try:
    data = pd.read_csv('Dataset_di_prova/public_health.csv')
except FileNotFoundError:
    print("File public_health.csv non trovato.")
    sys.exit()

data_filtrati = data[data['Country Name'].str.startswith('A')]

anni = ['Country Name', '1995', '1996', '1997', '1998', '1999']
data_selezionati = data_filtrati[anni]

print(data_selezionati)
```

# 25) Calcola la media dei valori in ogni colonna di un DataFrame.

```python
import numpy as np
import pandas as pd
import sys

try:
    data = pd.read_csv('Dataset_di_prova/public_health.csv')
except FileNotFoundError:
    print("File dataset_climatico.csv non trovato.")
    sys.exit()

media_valori_colonne = np.round(data.mean(numeric_only=True), 3)
media_valori_colonne
```
