# 26)  Crea un grafico a linee di una serie temporale con Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

mesi = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio', 'Giugno', 'Luglio', 'Agosto', 'Settembre', 'Ottobre', 'Novembre', 'Dicembre']
temperatura_media = [-4, 0, 7, 11, 16, 20, 24, 23, 19, 13, 9, 4]

df = pd.DataFrame({
    'Mese': mesi,
    'Temperatura': temperatura_media
})

df.set_index('Mese', inplace=True)

plt.figure(figsize=(10, 5)) 
plt.plot(df.index, df['Temperatura'], marker='o', linestyle='-', color='r')
plt.title('Temperature medie mensili') 
plt.xlabel('Mesi') 
plt.ylabel('Temperatura (°C)') 
plt.xticks(rotation=45) 
plt.grid(True) 

plt.show()
```

# 27) Crea un istogramma di una distribuzione di valori con Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
valori = np.random.randint(1, 101, 100) 

plt.hist(valori, bins=10, color='lightgreen', edgecolor='black')
plt.title('Istogramma della Distribuzione Casuale di 100 Valori')
plt.xlabel('Valore')
plt.ylabel('Frequenza')


plt.show()

```

# 28)  Crea un grafico a barre per rappresentare i dati di un DataFrame.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sopravvissuti_per_classe = titanic.groupby('class')['survived'].sum().reset_index()

plt.figure(figsize=(8, 6))  
sns.barplot(data=sopravvissuti_per_classe, x='class', y='survived', palette='Set2')
plt.title('Numero di Passeggeri Sopravvissuti per Classe')
plt.xlabel('Classe')
plt.ylabel('Numero di Sopravvissuti')

plt.show()
```

# 29) Crea un diagramma a dispersione (scatter plot) con due variabili di un DataFrame.
Il grafico a dispersione mostra quindi la relazione tra l'importo totale del conto e l'ammontare della mancia per i clienti del ristorante. 
Inoltre, i punti sono colorati in base al genere dei clienti, consentendo di vedere come il genere influisce sulla relazione tra le variabili.
Ad esempio, potresti notare se ci sono differenze nei comportamenti di mancia tra clienti maschi e femmine.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 6)) 
scatter = sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', palette='Set1')
plt.title('Diagramma a Dispersione di un ristorante differenziato tra conto e mance')
plt.xlabel('Conto totale' )
plt.ylabel('Mancia')
scatter.legend(title='Genere', labels=['Maschio', 'Femmina'])

plt.show()
```

# 30) Crea un box plot per analizzare la distribuzione in un DataFrame.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8, 6))  
sns.boxplot(data=titanic, x='age',  color='orange')
plt.title('Box Plot dell\'Età dei Passeggeri del Titanic')
plt.xlabel('Età')

plt.show()
```
